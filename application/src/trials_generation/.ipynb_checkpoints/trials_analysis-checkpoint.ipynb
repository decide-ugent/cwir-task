#### Helper functions

```python
def VisualizeGamble(lot_0_val, lot_0_prob, lot_1_val, lot_1_prob):
    """Function to visualize gambles
    """
    
    fig, ax = plt.subplots()

    # Define square properties
    square_size = 4  # Size of each square

    # Create two squares
    for i in range(2):
        x = i * (square_size + 2)  # Adjust spacing between squares to 2 units
        y = 0
        square = plt.Rectangle((x, y), square_size, square_size, fill=False, color='black')
        ax.add_patch(square)
        
        # Determine which lot values to use
        if i == 0:
            val = lot_0_val
            prob = lot_0_prob
        elif i == 1:
            val = lot_1_val
            prob = lot_1_prob
        
        # Position the text in the center of each square
        ax.text(x + square_size / 2, y + square_size / 2 - 1, f'{int(val)}$', fontsize=40, ha='center', va='center')
        ax.text(x + square_size / 2, y + square_size / 2 + 1, f'{int(prob)}%', fontsize=40, ha='center', va='center')

    # Set axis limits and labels
    ax.set_xlim(-0.2, 2 * (square_size + 2))
    ax.set_ylim(-0.2, square_size)
    ax.axis('off')

    # Display the plot
    plt.show()
```

## Aanalysis of potential rewards

#### Test average reward of a random agent

```python
n_random_agents = 6000

random_agents_rewards = np.zeros(n_random_agents)
```

```python
for i in tqdm(range(n_random_agents)):
    chosen_lotteries = np.zeros(n_trials)
    
    #random agent plays the game
    for j in range(n_trials):
        chosen_lotteries[j] = random.choice([0, 1])
        
    # List of integers from 1 to n_trials
    gamble_numbers = list(range(1, n_trials+1))

    # Draw 8 random lotteries (without replacement)
    random_draw = random.sample(gamble_numbers, 8)
    
    count = 0
    
    #Play each randomly drawn lottery from the bag of selected lotteries
    for gamble_num in random_draw:
        chosen_lot = chosen_lotteries[gamble_num-1]
        current_gamble = gambles[gambles['GambleNumber'] == gamble_num] 
        random_num = random.random()
        
        if chosen_lot == 0:
            lot_0_val = current_gamble['lot_0_val'].iloc[0] 
            lot_0_prob = current_gamble['lot_0_prob'].iloc[0] 
            
            if random_num < lot_0_prob:
                count += lot_0_val
            
        else:
            lot_1_val = current_gamble['lot_1_val'].iloc[0] 
            lot_1_prob = current_gamble['lot_1_prob'].iloc[0] 
            if random_num < lot_1_prob:
                count += lot_1_val
                
    random_agents_rewards[i] = count
            
        

```

```python
%matplotlib inline

plt.hist(random_agents_rewards)
plt.ylabel("Frequency")
plt.xlabel("Final number of points obtained")
plt.title("Final number of points obtained for random agent")
plt.show()
```

```python
print("Mean reward of random agent: "+str(np.mean(random_agents_rewards)))
```

#### Test average reward of "safe" agent

```python
n_safe_agents = 6000

safe_agents_rewards = np.zeros(n_safe_agents)
```

```python
for i in tqdm(range(n_safe_agents)):
    chosen_lotteries = np.zeros(n_trials)
    
    #safe agent plays the game
    for j in range(n_trials):
        
        current_gamble = gambles[gambles['GambleNumber'] == j+1] 
        lot_0_prob = current_gamble['lot_0_prob'].iloc[0]
        lot_1_prob = current_gamble['lot_1_prob'].iloc[0]
        
        if lot_0_prob >= lot_1_prob:
            chosen_lotteries[j] = 0
        else:
            chosen_lotteries[j] = 1
            
                   
    # List of integers from 1 to n_trials
    gamble_numbers = list(range(1, n_trials+1))

    # Draw 8 random lotteries (without replacement)
    random_draw = random.sample(gamble_numbers, 8)
    
    count = 0
    
    #Play each randomly drawn lottery from the bag of selected lotteries
    for gamble_num in random_draw:
        chosen_lot = chosen_lotteries[gamble_num-1]
        current_gamble = gambles[gambles['GambleNumber'] == gamble_num] 
        random_num = random.random()
        
        if chosen_lot == 0:
            lot_0_val = current_gamble['lot_0_val'].iloc[0] 
            lot_0_prob = current_gamble['lot_0_prob'].iloc[0] 
            
            if random_num < lot_0_prob:
                count += lot_0_val
            
        else:
            lot_1_val = current_gamble['lot_1_val'].iloc[0] 
            lot_1_prob = current_gamble['lot_1_prob'].iloc[0] 
            if random_num < lot_1_prob:
                count += lot_1_val
                
    safe_agents_rewards[i] = count
```

```python
%matplotlib inline

plt.hist(safe_agents_rewards)
plt.ylabel("Frequency")
plt.xlabel("Final number of points obtained")
plt.title("Final number of points obtained for safe agent")
plt.show()
```

```python
print("Mean reward of safe agent: "+str(np.mean(safe_agents_rewards)))
```

#### Test average reward of "optimal" agent

```python
n_optimal_agents = 6000

optimal_agents_rewards = np.zeros(n_optimal_agents)
```

```python
for i in tqdm(range(n_optimal_agents)):
    chosen_lotteries = np.zeros(n_trials)
    
    #Optimal agent plays the game
    for j in range(n_trials):
        
        current_gamble = gambles[gambles['GambleNumber'] == j+1] 
        lot_0_ev = current_gamble['lot_0_ev'].iloc[0]
        lot_1_ev = current_gamble['lot_1_ev'].iloc[0]
        
        if lot_0_ev >= lot_1_ev:
            chosen_lotteries[j] = 0
        else:
            chosen_lotteries[j] = 1
            
                   
    # List of integers from 1 to n_trials
    gamble_numbers = list(range(1, n_trials+1))

    # Draw 8 random lotteries (without replacement)
    random_draw = random.sample(gamble_numbers, 8)
    
    count = 0
    
    #Play each randomly drawn lottery from the bag of selected lotteries
    for gamble_num in random_draw:
        chosen_lot = chosen_lotteries[gamble_num-1]
        current_gamble = gambles[gambles['GambleNumber'] == gamble_num] 
        random_num = random.random()
        
        if chosen_lot == 0:
            lot_0_val = current_gamble['lot_0_val'].iloc[0] 
            lot_0_prob = current_gamble['lot_0_prob'].iloc[0] 
            
            if random_num < lot_0_prob:
                count += lot_0_val
            
        else:
            lot_1_val = current_gamble['lot_1_val'].iloc[0] 
            lot_1_prob = current_gamble['lot_1_prob'].iloc[0] 
            if random_num < lot_1_prob:
                count += lot_1_val
                
    optimal_agents_rewards[i] = count
```

```python
%matplotlib inline

plt.hist(optimal_agents_rewards)
plt.ylabel("Frequency")
plt.xlabel("Final number of points obtained")
plt.title("Final number of points obtained for optimal agent")
plt.show()
```

```python
print("Mean reward of optimal agent: "+str(np.mean(optimal_agents_rewards)))
```

#### Calculate the maximum reward that you can get

```python
max_rewards_list = np.zeros(n_trials)

for i in range(n_trials):
    current_gamble = gambles[gambles['GambleNumber'] == i+1]
    lot_0_val = current_gamble['lot_0_val'].iloc[0]
    lot_1_val = current_gamble['lot_1_val'].iloc[0]
    
    max_out = max([lot_0_val,lot_1_val])
    max_rewards_list[i] = max_out
```

```python
max_rewards_list_sorted = np.sort(max_rewards_list)[::-1]

first_eight_elements = max_rewards_list_sorted[:8]
print("Max possible reward: "+ str(np.sum(first_eight_elements)))
```
