```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
probabilities = np.genfromtxt('gambles_percentages.csv', delimiter=',')/100
points_values = np.genfromtxt('gambles_points_values.csv', delimiter=',')
```

```python
numTrials = len(probabilities)
```

```python
# Each row corresponds to a gamble, each column is the expected value of a lottery
expected_values = np.zeros((numTrials,3)) 

#Array where first column element is the lottery number with the highest expected value of the gamble.
#(i.e. if 2, then lottery 2 has highest expected value). The second column is the expected value of the best lottery.
best_lotteries = np.zeros((numTrials,2))

# Each row corresponds to a gamble, the first column is D1 (expected value difference between best and second best lotteries)
#The second column is D2 (expected value difference between best and worst lotteries)
expected_value_distances = np.zeros((numTrials,2))
```

```python
for i in range(numTrials):
    
    #Calculate expected values of each lottery for each gamble
    expected_values[i,0] = points_values[i][0]*(1- probabilities[i][0]) + points_values[i][1]*probabilities[i][0]
    expected_values[i,1] = points_values[i][2]*(1- probabilities[i][1]) + points_values[i][3]*probabilities[i][1]
    expected_values[i,2] = points_values[i][4]*(1- probabilities[i][2]) + points_values[i][5]*probabilities[i][2]
  
    max_expected_val = max(expected_values[i,0],expected_values[i,1],expected_values[i,2])
    best_lotteries[i,1] = max_expected_val
    
    #If the best lottery is lottery 1:
    if max_expected_val == expected_values[i,0]:
        best_lotteries[i,0] = 1
        
        #If lottery 2 is the second best
        if expected_values[i,1] >= expected_values[i,2]:
            #D1 is equal to the difference in expected value between lotteries 1 and 2
            expected_value_distances[i,0] = expected_values[i,0] - expected_values[i,1]
            #D2 is equal to the difference in expected value between lotteries 1 and 3
            expected_value_distances[i,1] = expected_values[i,0] - expected_values[i,2]
        #If lottery 3 is the second best   
        else:
            #D1 is equal to the difference in expected value between lotteries 1 and 3
            expected_value_distances[i,0] = expected_values[i,0] - expected_values[i,2]
            #D2 is equal to the difference in expected value between lotteries 1 and 2
            expected_value_distances[i,1] = expected_values[i,0] - expected_values[i,1]
                  
    #If the best lottery is lottery 2:
    if max_expected_val == expected_values[i,1]:
        best_lotteries[i,0] = 2
        
        #If lottery 1 is the second best
        if expected_values[i,0] >= expected_values[i,2]:
            #D1 is equal to the difference in expected value between lotteries 2 and 1
            expected_value_distances[i,0] = expected_values[i,1] - expected_values[i,0]
            #D2 is equal to the difference in expected value between lotteries 2 and 3
            expected_value_distances[i,1] = expected_values[i,1] - expected_values[i,2]
        
        #If lottery 3 is the second best   
        else:
            #D1 is equal to the difference in expected value between lotteries 2 and 3
            expected_value_distances[i,0] = expected_values[i,1] - expected_values[i,2]
            #D2 is equal to the difference in expected value between lotteries 2 and 1
            expected_value_distances[i,1] = expected_values[i,1] - expected_values[i,0]        
        
    #If the best lottery is lottery 3:
    if max_expected_val == expected_values[i,2]:
        best_lotteries[i,0] = 3
        
        #If lottery 1 is the second best
        if expected_values[i,0] >= expected_values[i,1]:
            #D1 is equal to the difference in expected value between lotteries 3 and 1
            expected_value_distances[i,0] = expected_values[i,2] - expected_values[i,0]
            #D2 is equal to the difference in expected value between lotteries 3 and 2
            expected_value_distances[i,1] = expected_values[i,2] - expected_values[i,1]
        #If lottery 2 is the second best   
        else:
            #D1 is equal to the difference in expected value between lotteries 3 and 2
            expected_value_distances[i,0] = expected_values[i,2] - expected_values[i,1]
            #D2 is equal to the difference in expected value between lotteries 3 and 1
            expected_value_distances[i,1] = expected_values[i,2] - expected_values[i,0]
        
    if expected_values[i,0] == expected_values[i,1]:
        print(f"For gamble {i}, lotteries 1 and 2 have the same expected values")
    if expected_values[i,0] == expected_values[i,2]:
        print(f"For gamble {i}, lotteries 1 and 3 have the same expected values")
    if expected_values[i,1] == expected_values[i,2]:
        print(f"For gamble {i}, lotteries 2 and 3 have the same expected values")
```

```python
# Plot Gambles in D1/D2 plain

plt.figure(figsize=(7, 7))
# Set the aspect ratio to be equal, making the plot square
plt.gca().set_aspect('equal', adjustable='box')

# Set the x and y axis limits from 0 to 20
plt.xlim(0, 20)
plt.ylim(0, 20)

# Set the x and y axis ticks to integers from 0 to 20
plt.xticks(range(21))
plt.yticks(range(21))

plt.grid(True, linestyle='--', alpha=0.6)
# Add labels and title

plt.xlabel('D1')
plt.ylabel('D2')
plt.title('Gambles in D1/D2 plain')
plt.plot(expected_value_distances[:, 0], expected_value_distances[:, 1], 'o', label='Points', markersize=8, color='b')
plt.plot([0, 20], [0, 20], linestyle='--', color='r')

# Show the plot
plt.show()

#Remember! D1 should always be smaller (or equal) than D2
```

```python
# Create a histogram with custom bins
bins = list(range(-10, 11))  # Custom bin edges from -10 to 10
plt.hist(best_lotteries[:,1], bins=bins, color='g', alpha=0.7, rwidth=0.85, label='Histogram')

# Add labels and title
plt.xlabel('Point values')
plt.ylabel('Frequency')
plt.title('Expected values of best options')


# Show the plot
plt.show()


#It is reasonable that the expected values of the best options tend to be on the positive side of the graph
```
